```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
import numpy as np # Import numpy
import imageio # import imagio

pipe = DiffusionPipeline.from_pretrained(
    "cerspense/zeroscope_v2_576w",
    torch_dtype=torch.float16
)

pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()
pipe.vae.enable_slicing()

prompt = "A cute cat happily playing with a small red ball"
result = pipe(prompt, num_inference_steps=20, height=320, width=576, num_frames=16)

video_frames = result.frames[0]       # remove batch dimension
video_frames = (video_frames * 255).astype(np.uint8)

output_video_path = "output.mp4"
fps = 8

with imageio.get_writer(output_video_path, fps=fps) as writer:
    for frame in video_frames:
        writer.append_data(frame)

output_video_path
```
