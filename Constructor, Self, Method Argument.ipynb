```python
class computer():
    
    def __init__(self,cpu,ram, price, company): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object 
        self.ram = ram
        self.price = price
        self.company = company
        
    def config(self):
        print('config is', self.cpu, self.price, self.company)
        
com1 = computer('i5', 8, 5000, 'hp')
com2 = computer('i7', 16, 80000, 'lenovo')

com1.config()
com2.config()

#print(id(com1))
#print(id(com2))

        
```
